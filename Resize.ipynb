```python
import os
import numpy as np
import cv2
import random
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from PIL import Image
sns.set()
import glob
```

```python
WIDTH = 128
HEIGHT = 128
```

```python
source = 'Horror'
files = os.listdir(source)
```

```python
plt.subplot(1,3,1)
g = cv2.imread('Horror' + '/' + files[5])
plt.imshow(g)

plt.subplot(1,3,2)
g = cv2.imread('Horror' + '/' + files[4])
plt.imshow(g)

plt.subplot(1,3,3)
g = cv2.imread('Horror' + '/' + files[3])
plt.imshow(g)
```

```python
for i in files:
    full_size_image = cv2.imread('Horror' + '/' + i)
    j = cv2.resize(full_size_image, (WIDTH,HEIGHT), interpolation=cv2.INTER_CUBIC)
    np.save('Resized_Horror' + '/' + i + '.npy', j)
```

```python
source = 'Resized_Horror'
files = os.listdir(source)
```

```python
#g = cv2.imread('Resized_Horror' + '/' + files[5])
g = np.load('Resized_Horror' + '/' + files[5])
plt.imshow(g)
```
